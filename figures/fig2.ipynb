```python
import numpy as np
import matplotlib.pyplot as plt 
import pickle
from mpl_toolkits.mplot3d import Axes3D
```

```python
time = 7100
xy_coord_folder = "../model/layout_healthy/dendritic-tree-coordinate"
z_coord_folder = "../model/layout_healthy/z-coordinate"
data_folder = "../data/fig3"
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,5))

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.01, 0.7])

for i, cell_type in enumerate(["cr"]):
    x_y = np.load("{}/{}.npy".format(xy_coord_folder, cell_type.upper()))
    x, y = x_y[0], x_y[1]
    with open('{}/data_{}.pickle'.format(data_folder, cell_type), 'rb') as f:
        color = pickle.load(f)["V_value"][time]
        print(color.min(), color.max())
    sc = ax.scatter(x, y, s=10, c=list(color), cmap='plasma', vmin=-51, vmax=-46)
    ax.set_aspect('equal')
    ax.set_title("{}\nmin {:.2f} mV, max {:.2f} mV".format(cell_type, color.min(), color.max()), fontsize=15)
    if i == 0:
        cbar = fig.colorbar(sc, cax=cbar_ax)
        cbar.set_label('membrane potential (mV)', rotation=270, labelpad=30, fontsize=15)
```

```python
t = time
data_dictionary = {}
for cell_type in ["cr", "hrz", "bp_on", "bp_off", "am_wf_on", "am_wf_off", "am_nf_on", "gl_on", "gl_off"]:
    x_y = np.load("{}/{}.npy".format(xy_coord_folder, cell_type.upper()))
    x, y = x_y[0], x_y[1]
    z = np.load("{}/{}.npy".format(z_coord_folder, cell_type.upper()))[0]
    with open('{}/data_{}.pickle'.format(data_folder, cell_type), 'rb') as f:
        color_data = pickle.load(f)["V_value"][t]
    data_list = [x, y, z, color_data]
    data_dictionary[cell_type] = data_list
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')

ax.set_xlim3d(-150,150)
ax.set_ylim3d(-150,150)
ax.set_zlim3d(0,200)
ax.invert_zaxis()
            
for key in data_dictionary:

    x = data_dictionary[key][0]
    y = data_dictionary[key][1]
    z = data_dictionary[key][2]
    color_data = data_dictionary[key][3]
    
    if key == "cr":
        vmin = -51
        vmax = -46
        sc = ax.scatter(x, y, z, s=10, c=list(color_data), cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
    if key == "hrz":
        vmin = -56
        vmax = -46
        sc = ax.scatter(x, y, z, s=10, c=list(color_data), cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
    if key == "bp_on":
        vmin = -36
        vmax = -30
        sc = ax.scatter(x, y, z, s=10, c=list(color_data), cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
    if key == "am_wf_on":
        vmin = -43
        vmax = -39
        sc = ax.scatter(x, y, z, s=10, c=list(color_data), cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
    if key == "am_nf_on":
        vmin = -48
        vmax = -46
        sc = ax.scatter(x, y, z, s=10, c=list(color_data), cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
    if key == "gl_on":
        vmin = -70
        vmax = -0
        size = np.ones(color_data.shape)*10
        size[color_data>0] = 100
        sc = ax.scatter(x[color_data<=0], y[color_data<=0], z[color_data<=0], s=size[color_data<=0], c=list(color_data[color_data<=0]), cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
        sc = ax.scatter(x[color_data>0], y[color_data>0], z[color_data>0], s=size[color_data>0], c=list(color_data[color_data>0]), edgecolors='black', cmap='plasma', vmin=vmin, vmax=vmax, alpha=0.3)
        
ax.view_init(elev=10, azim=-60)
plt.title("ON Pathway")
```
