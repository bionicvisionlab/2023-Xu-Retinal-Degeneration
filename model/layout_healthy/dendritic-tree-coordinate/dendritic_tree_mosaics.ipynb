**Notes**

Currently, the mosaics are generated by cropping a (very) large hexagonal lattice. As long as the center $\mathbf{p}$ is within the limits, that center is included in the dendritic tree mosaics. This is inefficient due to the for-loops and the extra unnecessary calculations.

But the limits of $i$ and $j$ can be calculated beforehand. $L$ is the cropping limit, which in the paper is 150 $\mu$m.

$$-\frac{L}{\lambda} \leq i + j \leq \frac{L}{\lambda},$$
$$-\frac{L}{\sqrt{3}\lambda} \leq i - j \leq \frac{L}{\sqrt{3}\lambda}.$$

Then we have

$$ - \frac{(1+\sqrt{3})L}{2\sqrt{3}\lambda} \leq i \leq \frac{(1+\sqrt{3})L}{2\sqrt{3}\lambda},$$
$$ \max(-i + \frac{L}{\lambda}, i - \frac{L}{\sqrt{3}\lambda}) \leq j \leq \min(-i + \frac{L}{\lambda}, i + \frac{L}{\sqrt{3}\lambda}) \text{ (for each fixed }i).$$

From $i$ and $j$ we will then be able to calculate $\mathbf{p}$ without the extra unnecessary calculations.

But at the boundaries we would still have to check whether $\mathbf{p}$ is within the cropping limit due to the unknown (normal) Gaussian noise.

For-loops will still be used in this scheme though.

So we will keep the this inefficient method for now, because it's not that slow.

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10,10)
```

```python
np.random.seed(0)
```

```python
# Section 2.3.2: 300*300*210 grid
LIMIT = 150
```

```python
# Table 2
lambda_cr = 2.5
lambda_hrz = 7.0
lambda_bp = 3.85
lambda_am_wf = 8.0
lambda_am_nf = 6.0
lambda_gl = 6.0
```

**Cone photoreceptors**

```python
xs = []
ys = []
for i in range(-100,101):
    for j in range(-100,101):
        p_x = (i+j) * lambda_cr + np.random.normal()
        p_y = np.sqrt(3) * (i-j) * lambda_cr + np.random.normal()
        if p_x > LIMIT or p_x < -LIMIT or p_y > LIMIT or p_y < -LIMIT:
            continue
        xs.append(p_x)
        ys.append(p_y)
plt.xlim(-150,150)
plt.ylim(-150,150)
plt.axis('square')
plt.scatter(xs, ys)
plt.savefig("dendritic-tree-coordinate/CR.jpg")
cr_coordinates = np.array([xs, ys])
np.save("dendritic-tree-coordinate/CR.npy", cr_coordinates)
```
